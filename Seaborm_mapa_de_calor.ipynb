```python
import seaborn as sns
import pandas as pd
import numpy as np
# Cargar el conjunto de datos "Tips" desde seaborn
tips = sns.load_dataset('tips')

# Crear un nuevo DataFrame para análisis adicional
tips_extended = tips.copy()

# Calcular la propina como porcentaje del total de la cuenta
tips_extended['tip_percentage'] = tips_extended['tip'] / tips_extended['total_bill'] * 100

# Agregar una columna para categorizar la propina como baja, media o alta
tips_extended['tip_category'] = np.where(tips_extended['tip_percentage'] < 15, 'Baja',np.where(tips_extended['tip_percentage'] < 20, 'Media', 'Alta'))

# Crear un gráfico de dispersión con categorías de propina utilizando seaborn
sns.scatterplot(x='total_bill', y='tip', hue='tip_category', style='time', size='size',sizes=(50, 200), alpha=0.7, data=tips_extended)

     # Calcular la matriz de correlación corr_matrix = tips.corr()  # Crear el mapa de calor utilizando seaborn sns.heatmap(corr_matrix, annot=True, cmap='coolwarm') 
```
